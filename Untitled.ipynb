```python
# Repository Migration Verification

import os
from pathlib import Path
import subprocess

def verify_directory_structure():
    """Verify the expected directory structure exists"""
    
    expected_structure = {
        'projects': [
            '01-mcp-basics',
            '02-github-integration',
            '03-travel-assistant'
        ],
        'docs': [
            'learning-notes',
            'progress',
            'references'
        ],
        'experiments': [],
        'resources': [],
        'scripts': []
    }
    
    print("ğŸ” Verifying Directory Structure")
    print("=" * 40)
    
    all_good = True
    
    for main_dir, sub_dirs in expected_structure.items():
        if os.path.exists(main_dir):
            print(f"âœ… {main_dir}/ exists")
            
            if sub_dirs:
                for sub_dir in sub_dirs:
                    full_path = os.path.join(main_dir, sub_dir)
                    if os.path.exists(full_path):
                        print(f"  âœ… {main_dir}/{sub_dir}/ exists")
                    else:
                        print(f"  âŒ {main_dir}/{sub_dir}/ missing")
                        all_good = False
        else:
            print(f"âŒ {main_dir}/ missing")
            all_good = False
    
    return all_good

def verify_files_moved():
    """Verify specific files are in expected locations"""
    
    expected_files = {
        'projects/01-mcp-basics/mcp-reference-notebook.md': 'MCP Reference Notebook',
        'projects/03-travel-assistant/travel_assistant_mini.ipynb': 'Travel Assistant Mini',
        'projects/03-travel-assistant/travel_assistant_complete.ipynb': 'Travel Assistant Complete',
        'projects/02-github-integration/github_mcp.ipynb': 'GitHub MCP Integration'
    }
    
    print("\nğŸ“ Verifying File Locations")
    print("=" * 40)
    
    all_files_found = True
    
    for file_path, description in expected_files.items():
        if os.path.exists(file_path):
            print(f"âœ… {description}: {file_path}")
        else:
            print(f"âŒ {description}: {file_path} (not found)")
            all_files_found = False
    
    return all_files_found

def verify_git_status():
    """Check git status for any uncommitted changes"""
    
    print("\nğŸ“‹ Checking Git Status")
    print("=" * 40)
    
    try:
        result = subprocess.run(['git', 'status', '--porcelain'], 
                              capture_output=True, text=True, check=True)
        
        if result.stdout.strip():
            print("âš ï¸ Uncommitted changes found:")
            print(result.stdout)
            return False
        else:
            print("âœ… Working directory is clean")
            return True
    except subprocess.CalledProcessError as e:
        print(f"Error checking git status: {e}")
        return False

def check_git_tags():
    """Check for migration tag"""
    
    print("\nğŸ·ï¸ Checking Git Tags")
    print("=" * 40)
    
    try:
        result = subprocess.run(['git', 'tag'], 
                              capture_output=True, text=True, check=True)
        
        tags = result.stdout.strip().split('\n')
        if 'monorepo-migration' in tags:
            print("âœ… Migration tag found: monorepo-migration")
            return True
        else:
            print("âŒ Migration tag not found")
            return False
    except subprocess.CalledProcessError as e:
        print(f"Error checking git tags: {e}")
        return False

def list_remaining_files():
    """List any files in the root directory"""
    
    print("\nğŸ“‹ Files in Root Directory")
    print("=" * 40)
    
    root_files = []
    for item in os.listdir('.'):
        if os.path.isfile(item) and not item.startswith('.'):
            root_files.append(item)
    
    if root_files:
        print("Files still in root directory:")
        for file in root_files:
            print(f"  - {file}")
    else:
        print("âœ… No unexpected files in root directory")
    
    return root_files

def provide_next_steps(issues_found):
    """Provide guidance for resolving any issues"""
    
    print("\nğŸ¯ Next Steps")
    print("=" * 40)
    
    if not issues_found:
        print("âœ… Migration completed successfully!")
        print("\nRecommended next steps:")
        print("1. Push to remote: git push origin main --tags")
        print("2. Start using your organized repository")
        print("3. Create a feature branch for your next project")
    else:
        print("âš ï¸ Some issues found. Here's what to do:")
        
        if 'uncommitted_changes' in issues_found:
            print("\n1. Commit any remaining changes:")
            print("   git add .")
            print("   git commit -m 'fix: Complete migration cleanup'")
        
        if 'files_not_moved' in issues_found:
            print("\n2. Move remaining files manually:")
            print("   Use: git mv <file> <destination>/")
        
        if 'missing_directories' in issues_found:
            print("\n3. Create missing directories:")
            print("   mkdir -p <directory-path>")
        
        print("\n4. Once fixed, push to remote:")
        print("   git push origin main --tags")

# Run verification
def run_complete_verification():
    """Run all verification checks"""
    
    print("ğŸ” Repository Migration Verification")
    print("=" * 40)
    print()
    
    issues_found = []
    
    # Check directory structure
    if not verify_directory_structure():
        issues_found.append('missing_directories')
    
    # Check file locations
    if not verify_files_moved():
        issues_found.append('files_not_moved')
    
    # Check git status
    if not verify_git_status():
        issues_found.append('uncommitted_changes')
    
    # Check for migration tag
    check_git_tags()
    
    # List remaining files
    remaining_files = list_remaining_files()
    if remaining_files:
        issues_found.append('remaining_files')
    
    # Provide next steps
    provide_next_steps(issues_found)
    
    return len(issues_found) == 0

# Execute verification
if __name__ == "__main__":
    success = run_complete_verification()
    
    if success:
        print("\nğŸ‰ Repository migration verified successfully!")
    else:
        print("\nâš ï¸ Please address the issues above before proceeding.")
```

```python
# Fix repository directory structure

import subprocess
import os

def fix_repository_structure():
    """Move files to correct locations and clean up"""
    
    # List of commands to fix the structure
    commands = [
        # Move files from mcp-travel-assistant to correct locations
        "git mv projects/mcp-travel-assistant/mcp-reference-notebook.md projects/01-mcp-basics/",
        "git mv projects/mcp-travel-assistant/travel_assistant_complete.ipynb projects/03-travel-assistant/",
        "git mv projects/mcp-travel-assistant/travel_assistant_mini.ipynb projects/03-travel-assistant/",
        "git mv projects/mcp-travel-assistant/README.md projects/03-travel-assistant/",
        
        # Remove empty mcp-travel-assistant directory
        "rmdir projects/mcp-travel-assistant",
        
        # Move Untitled.ipynb to experiments
        "git mv Untitled.ipynb experiments/",
        
        # Remove extra directories that shouldn't exist
        "rm -rf git_examples",
        "rm -rf learning_notes",
        "rm -rf mcp_fundamentals"
    ]
    
    print("ğŸ”§ Fixing repository structure...")
    print("=" * 40)
    
    for cmd in commands:
        try:
            if cmd.startswith('rm '):
                # For rm commands, use os.system to handle wildcards and directories
                os.system(cmd)
                print(f"âœ… {cmd}")
            else:
                # For git commands, use subprocess
                subprocess.run(cmd.split(), check=True)
                print(f"âœ… {cmd}")
        except subprocess.CalledProcessError as e:
            print(f"âš ï¸ {cmd} - {str(e)}")
        except Exception as e:
            print(f"âš ï¸ {cmd} - {str(e)}")
    
    # Commit all changes
    print("\nğŸ“ Committing changes...")
    try:
        subprocess.run(["git", "add", "."], check=True)
        subprocess.run(["git", "commit", "-m", "fix: Reorganize files to correct locations"], check=True)
        print("âœ… Changes committed")
    except subprocess.CalledProcessError as e:
        print(f"âš ï¸ Commit failed: {e}")
    
    # Show final status
    print("\nğŸ“‹ Final repository status:")
    subprocess.run(["git", "status"], check=False)

# Run the fix
fix_repository_structure()

# Verify the structure
print("\nâœ¨ Repository structure should now be:")
print("""
learn_git_and_mcp/
â”œâ”€â”€ projects/
â”‚   â”œâ”€â”€ 01-mcp-basics/
â”‚   â”‚   â”œâ”€â”€ README.md
â”‚   â”‚   â””â”€â”€ mcp-reference-notebook.md
â”‚   â”œâ”€â”€ 02-github-integration/
â”‚   â”‚   â””â”€â”€ README.md
â”‚   â””â”€â”€ 03-travel-assistant/
â”‚       â”œâ”€â”€ README.md
â”‚       â”œâ”€â”€ travel_assistant_complete.ipynb
â”‚       â””â”€â”€ travel_assistant_mini.ipynb
â”œâ”€â”€ docs/
â”œâ”€â”€ experiments/
â”œâ”€â”€ resources/
â”œâ”€â”€ scripts/
â””â”€â”€ Untitled.ipynb (moved to experiments/)
""")

print("\nğŸ¯ Next steps:")
print("1. Verify the structure looks correct")
print("2. Push to remote: git push origin main --tags")
print("3. Start using your organized repository!")
```
